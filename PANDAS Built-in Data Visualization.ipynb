# Plot Types
There are several plot types built-in to pandas, most of them statistical plots by nature:


```python
*df.plot.area
*df.plot.barh
*df.plot.density
*df.plot.hist
*df.plot.line
*df.plot.bar
*df.plot.box
*df.plot.hexbin
*df.plot.scatter
```

# Area

```python
df2.plot.area()
```

# Bar plot

```python
df2.plot.bar()
```

```python
df2.plot.barh()
```

```python
df2.plot.bar(stacked=True)
```

```python
df2.plot.pie(subplots=True)
```

# Histograms

```python
df1['A'].plot.hist(bins=40)   # bins help to see more data point
```

# Line plots

```python
df1['B'].plot.line(x=df1.index,y='B',figsize=(12,3),lw=2)
```

# Scatter Plots

```python
df1.plot.scatter(x='A',y='B')
```

```python
df1.plot.scatter(x='A',y='B',c='C', cmap='coolwarm') #cmap means colormap
```

```python
df1.plot.scatter(x='A',y='B',s=df1['C']*150)
```

# Boxplots

```python
df2.plot.box() # Can also pass a by= argument for groupby
```

# Haxagonal Bin plot
Useful for Bivariate Data, alternative to scatterplot:

```python
df=pd.DataFrame(np.random.randn(1000, 2), columns=['a','b'])
df.plot.hexbin(x='a',y='b',gridsize=25,cmap='Oranges')
```

# Kernel Density Estimation plot(KDE)
